# Collaborative Recommender system

```python
x=final_df.groupby('user_id').count()['rating']>=100
frequent_readers = x[x].index
final_df.shape
```

```python
filtered_rating =final_df[final_df['user_id'].isin(frequent_readers)]
```

```python
filtered_rating.groupby('title').count()['rating']>50
```

```python
y=filtered_rating.groupby('title').count()['rating']>50
```

```python
famous_movie = y[y].index
```

```python
final_ratings= filtered_rating[filtered_rating['title'].isin(famous_movie)]
```

```python
final_ratings
```

```python
final_ratings= final_ratings.drop_duplicates()
```

```python
final_ratings.shape
```

```python
final_ratings=final_ratings.pivot_table(index='title',columns='user_id',values='rating')
```

```python
final_ratings
```

```python
final_ratings.fillna(0,inplace=True)
```

```python
final_ratings
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_score = cosine_similarity(final_ratings)
```

```python
similarity_score.shape
```

```python
def recommend(movie_name):
    index=np.where(final_ratings.index==movie_name)[0][0]
    similar_movie = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)
    for i in similar_movie:
        print(final_ratings.index[i[0]])
```

# Will Recommend alternate movies Based on movies_name
